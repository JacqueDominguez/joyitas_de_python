```python
import pandas as pd
```

Creamos los Dataframes

```python
# Create Population DataFrame
df1 = pd.DataFrame({
    'Country': ['America', 'Indonesia', 'France', 'Italy'],
    'Location': ['New York', 'Jakarta', 'Paris','Rome'],
    'Population': [731800, 575030, 183305, 159863]
})

# Create Income DataFrame
df2 = pd.DataFrame({
    'Countries': ['America', 'America', 'Indonesia', 'India', 'France', 'Greece'],
    'Location': ['New York', 'Chicago', 'Jakarta', 'Mumbai', 'Paris', 'Yunani'],
    'Income': [1000, 1500, 1400, 1100, 900, 1200]

})
```

Combinamos ambas tablas con Full Outer Join y le creamos una nueva columna con el resultado

```python
merge = pd.merge(df1, df2 , how='outer', indicator='resultado',sort=True, left_on='Country', right_on='Countries')
```

Ahora el resultado lo pueden obsevar en la columna "RESULTADO"

* both = el dato está en los DOS data frames
* left_only = el dato sólo está en el DF1
* right_only = el dato sólo está en el DF2

Para ver las diferencias filtramos en la columna rsultado los elementos que SÓLO están en el DF1 y NO están en el DF2 - LEFT JOIN IF NULL

```python
merge[merge.resultado=='left_only']
```

Para ver las diferencias filtramos los elementos que SÓLO están en el DF2 y NO están en el DF1 - RIGHT JOIN IF NULL

```python
merge[merge.resultado=='right_only']
```

Vemos las columnas que coinciden - INNER JOIN

```python
merge[merge.resultado=='both']
```

```python
right = pd.merge(df1, df2 , how='right',sort=True, left_on='Country', right_on='Countries')
```
