```python
import pandas as pd
```

```python
student_info = [('George', 78, 'Australia'),
			('Micheal', 189, 'Germany'),
			('Oliva', 140, 'Malaysia'),
			('James', 95, 'Uganda'),
			('James', 95, 'Uganda'),
      ('Micheal', 200, 'Uganda'),
			('Oliva', 140, 'Malaysia'),
			('Elijah', 391, 'Japan'),
			('Chris', 167, 'China')
			]

df = pd.DataFrame(student_info,
				columns = ['Student_name', 'Student_id', 'Student_city'])


```

¿Hay duplicados? ¿Cuántos?

```python
df.duplicated().sum()
```

¿En una colunma o varias?

```python
df.Student_name.duplicated().sum() #Duplicados solo de la columna
```

```python
df.duplicated(subset=['Student_name','Student_city']).sum() #Duplicados considerando solo las 2 columnas
```

Hacer un nuevo dataframe solo con los duplicados del dataframe

```python
duplicate=df[df.duplicated(keep=False)] #Devuelve el registro original y sus repetidos
duplicate
```

Duplicados de 1 sola columna y los últimos registros

```python
new_duplicate = df[df.duplicated('Student_city',keep='last')]

new_duplicate
```

Eliminamos los duplicados

```python
df.drop_duplicates(keep='first',inplace=True) #Elimino los ultimos registros duplicados en el DF original quedando los primeros como válidos
df
```
